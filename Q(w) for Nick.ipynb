```python
'''This code finds the set Delta(winv) = \Delta_0^{Q(w)} given w. '''
import numpy as np
import numpy
from numpy.linalg import inv
from fractions import Fraction
from itertools import permutations #for permutations(range(n)):

# Notation setup

n = 5
print "\nG = SL(%d) \n" % n
''' Define basis beta and varpi '''
beta = numpy.zeros((n-1,n))
varpi = numpy.zeros((n-1,n))
for i in range(n-1):
    for j in range(n):
        if j == i:
            beta[i,j] = 1
        elif j == i + 1:
            beta[i,j] = -1
        else:
            beta[i,j] = 0
    for j in range(n):
        if j <= i:
            varpi[i,j] = (n-i-1)/float(n)
        else: 
            varpi[i,j] = -(i+1)/float(n)

def deltaof(winv):
    mylist = []
    for i in range(n-1):
        temp = np.zeros(n)
        for j in range(n):
            temp[j] = varpi[i][winv[j]]
        if not np.array_equal(temp, varpi[i]):
            mylist.append(i)
    return mylist

for w in permutations(range(n)):
    print "For w = ", w, " Delta(w) = ", deltaof(w)
```
